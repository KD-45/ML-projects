```python
features = data.loc[:, ['PetalLengthCm', 'PetalWidthCm']]
species = data.loc[:, 'Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
f_values = features.values
s_values = species.values
```

```python
X_test,X_train,y_test,y_train = train_test_split(f_values, s_values, test_size=0.3, random_state=45)
```

```python
X_test.shape,X_train.shape,y_test.shape,y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
iris_model = DecisionTreeClassifier()
iris_model.fit(X_train,y_train)
predictions = iris_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(classification_report(y_test, predictions))
```

```python
cm = confusion_matrix(y_test, predictions)
```

```python
cm_df = pd.DataFrame(cm,
                     index = ['SETOSA','VERSICOLR','VIRGINICA'], 
                     columns = ['SETOSA','VERSICOLR','VIRGINICA'])
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_classifier = DecisionTreeClassifier(criterion = 'gini')
decision_tree_classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
def plot_decision_tree(decision_tree_classifier):
    plt.figure(figsize=(16, 14))
    tree.plot_tree(decision_tree_classifier, fontsize = 15, feature_names = features.columns, 
                   class_names = ['Setosa', 'Versicolor', 'Virginica'], filled = True)
    plt.savefig('C:/Users/rehan/decision_tree.png')
```

```python
plot_decision_tree(decision_tree_classifier)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
print("Training Accuracy : {0:.2f} %".format(iris_model.score(X_train, y_train) * 100))
print("Testing Accuracy : {} %".format(iris_model.score(X_test, y_test) * 100))
```
